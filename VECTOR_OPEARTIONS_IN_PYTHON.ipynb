```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
#plotting a vector
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy',scale=  1,color='r')
plt.quiver(0,0,-3,-4, scale_units='xy', angles='xy',scale=  1,color='b')
plt.quiver(0,0,7,-2, scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#vector addition
vector_1=np.array([0,0,1,2])
vector_2=np.array([0,0,3,3])
sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=  1,color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy',scale=  1,color='b')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#vector suubtraction
vector_1=np.array([0,0,2,3])
vector_2=np.array([0,0,3,-2])
difference=vector_1-vector_2
print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=  1,color='r')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy',scale=  1,color='b')
plt.quiver(0,0,-1,5, scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#multiplying vector by scalar
vector_1=np.array([0,0,2,3])
scalar=2
product=vector_1*scalar
print(product)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=  1,color='r')
plt.quiver(0,0,-1,-1.5, scale_units='xy',angles='xy',scale=1,color='y')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
