```python
mean_vec2=np.array([1,1,1])
cov_mat2=np.array(([1,0,0],[0,1,0],[0,0,1]))
class2=np.random.multivariate_normal(mean_vec2,cov_mat2,100)
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o',markersize=8,color='orange',alpha=0.2,label='class1')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^',markersize=8,color='blue',alpha=0.2,label='class2')
plt.show()
```

```python
#move 3d data to 2d data
#combine all data
all_data=np.concatenate((class1,class2))
```

```python
pca=PCA(n_components=2)
transformed_data=pca.fit_transform(all_data)
```

```python
x_approx=pca.inverse_transform(transformed_data)
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(111,projection='3d')
ax.plot(x_approx[0:100,0],x_approx[0:100,1],x_approx[0:100,2],'o')
ax.plot(x_approx[100:200,0],x_approx[100:200,1],x_approx[100:200,2],'^')
plt.show()
```
